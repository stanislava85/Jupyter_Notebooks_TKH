```python
max_product = 0

for row in range(4):
    for column in range(4):
        for direction in ((0,1), (1,0), (1,1), (-1,-1)):
            try:
                product = 1
                for l in range(4):
                    product *= arr[row + direction[0]*l][column + direction[1]*l]
#                     print(row + direction[0]*l, column + direction[1]*l)
#                     print(arr[row + direction[0]*l][column + direction[1]*l])
                if product > max_product:
                    max_product = product
            except IndexError:
                raise Exception("Error")

print(max_product)
```
