```python
new_grid = grid.split()
lst = [int(char) for char in new_grid]
df = np.array(lst).reshape(20,20)
```

```python
def left_right(grid):
    maxi = 0
    for i in range(20):
        for j in range(16):
            num = grid[i][j] * grid[i][j+1] * grid[i][j+2] * grid[i][j+3]
            if maxi < num:
                maxi = num

    return maxi

def up_down(grid):
    maxi = 0
    for i in range(16):
        for j in range(20):
            num = grid[i][j] * grid[i+1][j] * grid[i+2][j] * grid[i+3][j]
            if maxi < num:
                maxi = num

    return maxi

def diag_back(grid):
    maxi = 0
    for i in range(16):
        for j in range(16):
            num = grid[i][j] * grid[i+1][j+1] * grid[i+2][j+2] * grid[i+3][j+3]
            if maxi < num:
                maxi = num

    return maxi

def diag_front(grid):
    maxi = 0
    for i in range(20):
        for j in range(3, 16):
            num = grid[i][j] * grid[i-1][j+1] * grid[i-2][j+2] * grid[i-3][j+3]
            if maxi < num:
                maxi = num

    return maxi

def solution(grid):
    grid20 = list()
    grid = list(map(int, grid.split()))
    for i in range(20):
        line = 20 * i
        grid20.append(grid[line:line+20])

    m1 = up_down(grid20)
    m2 = left_right(grid20)
    m3 = diag_back(grid20)
    m4 = diag_front(grid20)

    tot = max(m1, m2, m3, m4)
    return tot

tot = solution(grid)
print(tot)
```
