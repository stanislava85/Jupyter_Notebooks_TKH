# Exercise 1

We'd like to analyze the `SNAP_history_1969_2019.csv` dataset, however it's missing some data. Find the missing values. Fix incorrect entries as well.

Some helpful formulas:

$ \texttt{Total Costs} = \texttt{Total Benefits} + \texttt{Other Costs} $

$ \texttt{Total Benefits} = \texttt{Total Costs} - \texttt{Other Costs} $

$ \texttt{Other Costs} = \texttt{Total Costs} - \texttt{Total Benefits} $

$ \texttt{Average Participation} = \frac{\texttt{Total Benefits}}{\texttt{Average Benefits Per Person}}$

$ \texttt{Average Benefits Per Person} = \frac{\texttt{Total Benefits}}{\texttt{Average Participation}}$

$ \texttt{Total Benefits} = \texttt{Average Benefits Per Person} * \texttt{Average Participation}$

Note: `Average Participation` is in the thousands and all costs  are in the millions.

Hint: Think about how we can filter for specific columns that are null and apply a fix for those missing values.


```python
# calling it s_df to not confuse it with the others 
# and because `snap_df` is too much to constantly write
s_df = pd.read_csv('../datasets/SNAP_history_1969_2019.csv')
s_df
```

```python
s_df.info()
```

```python
s_df[s_df.isnull().any(axis=1)]
```

```python
#based on the formulas given above we'll start populating the missing data
#first the Total Cost(M) missing values
s_df.loc[s_df["Total Costs(M)"].isnull(),'Total Costs(M)'] = s_df["Total Benefits(M)"] + s_df["Other Costs"]
```

```python
#alternative solution to populate values for for Total Cost(M)
# s_df['"Total Costs(M)"'].fillna(s_df['Total Benefits(M)'] + s_df["Other Costs"], inplce=True)
```

```python
#populationg the missing values for Total Benefits(M)
s_df.loc[s_df["Total Benefits(M)"].isnull(), 'Total Benefits(M)'] = s_df["Total Costs(M)"] - s_df["Other Costs"]
```

```python
#s_df[s_df.isnull().any(axis=1)]    #checking what's left that's missing
```

```python
s_df["Average Participation"] = s_df["Average Participation"] / 1000
```

```python
#estimating missing values for Avegare Participation
s_df.loc[s_df["Average Participation"].isnull(), 'Average Participation'] = s_df['Total Benefits(M)'] / s_df['Average Benefit Per Person']
```

```python
s_df[s_df.isnull().any(axis=1)]
```

```python
#estimating the missing values for Average Benefir Per Person
s_df.loc[s_df["Average Benefit Per Person"].isnull(), 'Average Benefit Per Person'] = s_df['Total Benefits(M)'] / s_df['Average Participation']
```

```python
s_df[s_df.isnull().any(axis=1)]

#I will leave this one value for Fiscal Year empty as it is not correlated to anything else.
#in other case scenarion I could populate and empty value with an average value or a sequential one...
```

```python
# calling it g_df to not confuse it with the others 
# and because `grades_df` is too much to constantly write
g_df = pd.read_csv('../datasets/grades.csv')
g_df
```

```python
g_df.info()
```

```python
sns.pairplot(g_df)
plt.show()
```

```python
g_df.corr()
#based on those correlations it seems that the realtionship between the reading and writing score is linear
```

```python
plt.figure(figsize=(5,5))
sns.jointplot(x=g_df['reading score'].values, y=g_df['writing score'].values, height=10)
plt.suptitle("Scatter Plot of Reading/Writing Scores Relationship")
plt.ylabel('Writing Scores')
plt.xlabel('Reading Scores')
plt.show()
```

```python
plt.scatter(x='reading score', y='writing score', data=g_df, s=2.5, alpha=0.75)
plt.suptitle("Scatter Plot of Reading/Writing Scores Relationship")
plt.xlabel('Reading Scores')
plt.ylabel('Writing Scores')
plt.show()
```

```python
#I'd like to see if there is any logistic regression here that could help us classify the gender based on those 2 scores
```

```python
def convert_sex_to_binary(gender):
    return 1 if gender == 'male' else 0
```

```python
g_df['sex'] = g_df['gender'].apply(convert_sex_to_binary)
#applying the function and adding a new column
```

```python
g_df.head()
```

```python
#creating a new df for males and females
g_df0 = g_df[g_df['sex'] == 0]
```

```python
g_df1 = g_df[g_df['sex'] == 1]
```

```python
g_df0
```

```python
sns.pairplot(g_df)
plt.show()
#math and sex realtionship may have a way to fit a sigmoid line
```

```python
plt.scatter(x='reading score', y='writing score', data=g_df1, c='blue', s=2.5, alpha=0.75, label='M')
plt.scatter(x='reading score', y='writing score', data=g_df0, c='pink',s=2.5, alpha=0.75, label='F')
plt.suptitle("Scatter Plot of Reading/Writing Scores Relationship")
plt.xlabel('Reading Scores')
plt.ylabel('Writing Scores')
plt.legend(loc='best')
plt.show()
#this realtionship is linear.. no logistic regression happening here
```

```python
plt.scatter(x='math score', y='sex', data=g_df1, c='blue', s=2.5, alpha=0.75, label='M')
plt.scatter(x='math score', y='sex', data=g_df0, c='pink',s=2.5, alpha=0.75, label='F')
plt.suptitle("Scatter Plot of Math Scores and Sex Relationship")
plt.xlabel('math Scores')
plt.ylabel('Sex')
plt.legend(loc='best')
plt.show()
#this also doesn't seem like a good model to fit a Sigmoid Line; cannot predict Sex accurately based on the Math Score
```

```python
#maybe we could be able to cassify data based on the degree of the parent

g_df['parental level of education'].unique()
```

```python
def convert_edu_to_binary(edu):
    if edu == "bachelor's degree" or edu=="master's degree" or edu=="associate's degree":
        return 1
    else:
        return 0
```

```python
g_df['parental_degree'] = g_df['parental level of education'].apply(convert_edu_to_binary)
```

```python
g_df.head()
```

```python
sns.pairplot(g_df)
plt.show()

```

```python
sns.heatmap(g_df.corr())
```

```python
g_df3 = g_df[g_df['parental_degree'] == 0]
g_df4 = g_df[g_df['parental_degree'] == 1]
```

```python
plt.scatter(x='writing score', y='parental_degree', data=g_df3, c='blue', s=2.5, alpha=0.75, label='Non')
plt.scatter(x='writing score', y='parental_degree', data=g_df4, c='pink',s=2.5, alpha=0.75, label='Degree')
plt.suptitle("Scatter Plot of Parental Degree and Writing Score Relationship")
plt.xlabel('Writing Score')
plt.ylabel('Parental Degree')
plt.legend(loc='best')
plt.show()
#this one also doesn't seem like a good candidate for a logistic regression
```
