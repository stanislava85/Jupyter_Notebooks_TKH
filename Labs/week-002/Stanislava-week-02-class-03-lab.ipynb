<br> Bcause sets do not have repeating values like dictionaries, it may be useful to use them in a situation when we want to avoid repetition.<br> Let's say we are looking at data of consumers who completed a survey about which states they have lived in.
```python
students = [{"name":"Rob","student_id": 115, "favorite_colors":{"red","green"}},{"name":"Clara","student_id": 15, "favorite_colors":{"blue","red","green"}}, {"name":"Sally","student_id": 21, "favorite_colors":{"black"}}, {"name":"Jim","student_id": 33, "favorite_colors":{"blue","green"}}]
   
#print(type(students))

#splitting the first dict in the list - Rob
rob = (students[0])
print(rob)
print(type(rob)) 

#turning the Rob dict into a list
rob_1 = list(rob.items())    
print(rob_1)
print(type(rob_1))
#turning the colors dict into a list
rob_2 = list(rob_1[2][1])
print(rob_2)
print(type(rob_2))
#converting the colors list into a set
rob_s = set(rob_2)
print(rob_s)
print(type(rob_s))

#splitting the 2nd dict in the list - Clara
clara = students[1]
print(clara)
print(type(clara))
#turning the Clara dict into a list
clara_1 = list(clara.items())    
print(clara_1)
print(type(clara_1))
#turning the colors dict into a list
clara_2 = list(clara_1[2][1])
print(clara_2)
print(type(clara_2))
#converting the colors list into a set
clara_s = set(clara_2)
print(clara_s)
print(type(clara_s))


#splitting the 3rd dict in the list - Sally
sally = students[2]
print(sally)
print(type(sally))
#turning the Sally dict into a list
sally_1 = list(sally.items())    
print(sally_1)
print(type(sally_1))
#turning the colors dict into a list
sally_2 = list(sally_1[2][1])
print(sally_2)
print(type(sally_2))
#converting the colors list into a set
sally_s = set(sally_2)
print(sally_s)
print(type(sally_s))


#splitting the 4th dict in the list - Jim
jim = students[3]
print(jim)
print(type(jim))
#turning the Jim dict into a list
jim_1 = list(jim.items())    
print(jim_1)
print(type(jim_1))
#turning the colors dict into a list
jim_2 = list(jim_1[2][1])
print(jim_2)
print(type(jim_2))
#converting the colors list into a set
jim_s = set(jim_2)
print(jim_s)
print(type(jim_s))


#finding if there are intersections between each pair of sets
if rob_s.intersection(clara_s):   
    print("Rob and Clara")

if rob_s.intersection(sally_s):
    print("Rob and Sally")
    
if clara_s.intersection(sally_s):
    print("Clara and Sally")
    
if rob_s.intersection(jim_s):
    print("Rob and Jim")

if clara_s.intersection(jim_s):
    print("Clara and Jim")
    
if sally_s.intersection(jim_s):
    print("Sally and Jim")

```
