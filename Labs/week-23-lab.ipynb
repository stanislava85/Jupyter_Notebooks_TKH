# Exercise 1

Plot one chart using matplotlib (OOP-style) using the following data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/GOOG.csv')
df
```

```python
df.info()
```

```python
sns.pairplot(df)
```

```python
df1= df["Date"][0:20]
df2 = df["Close"][0:20]
df3 = df["Open"][0:20]
```

```python
fig, ax = plt.subplots(figsize=(16, 6))  # Create a figure and an axes.
ax.plot(df1, df2, label='Close') 
ax.plot(df1, df3, label='Open')  
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[0, 19], xlabel='Dates', ylabel='Values',
       title='Googles Opening and Closing Stats for 20 day range')
ax.legend()
plt.show()
```

```python
#or instead of saving into a new df refer to the columns
dates = tuple([i for i in df["Date"].value_counts().keys()])[:20]
type(dates)
```

```python
values = df[["Open", "High", "Low", "Close"]][:20]
type(values)
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

ax.plot(dates,values)
ax.set_xlabel("Date")
ax.set_ylabel("Values")
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[0, 19], title='Googles Opening and Closing Stats for 20 day range')
ax.legend(["Open", "High", "Low", "Close"])
plt.show()
```
